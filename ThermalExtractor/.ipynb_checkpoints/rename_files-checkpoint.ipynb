```python
import os
import glob
import re
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
#sanity check
original_images = []

for directory_path in glob.glob("C:\\Users\\desuo\\Desktop\\Projeto Masto Fidi\\images"):
    for img_path in glob.glob(os.path.join(directory_path, "*.png")):
        print(img_path)
        img = cv2.imread(img_path,  0)
        original_images.append(img)
        
```

```python
processed_images = []

for directory_path in glob.glob("C:\\Users\\desuo\\Desktop\\Projeto Masto Fidi\\masksmarques"):
    for mask_path in glob.glob(os.path.join(directory_path, "*.png")):
        print(mask_path)
        processed = cv2.imread(mask_path, 0)
        processed_images.append(processed)
```

```python
import random
image_number = random.randint(0, len(original_images)-1)
img = original_images[image_number]
mask = processed_images[image_number]

plt.figure(figsize=(12,8))
plt.subplot(221)
plt.title("original")
plt.imshow(img)
plt.subplot(222)
plt.title("processed")
plt.imshow(mask)
plt.show()
```
